# SARIMAX
When it comes to time series modelling, there is one model which should always be mentioned. Namely an *ARMA*. ARMA stands for **A**uto**r**egressive **m**oving **a**verage. That also already says which two classes make up an ARMA. 

In the following two secions we elaborate a bit what an Autoregressive (AR) and Moving Average (MA) process is and how they work. Afterwards we introduce the seasonality components, which adds the **S** in SARIMAX, before last but not least talking about the inclusion of e**x**egnous variables. 

## Moving Averages
In a moving average model, we use past forecasting errors in order to predict the current endogenous variable. Putting that in mathematical terms, a 1st order MA model would look the following way:
$$y_t = \phi_0 = \theta_1 \epsilon_{t-1} + \epsilon_t$$
In that equation we have two parameters, namely the intercept $\phi_0$ and the weighting parameter $\theta_1$. $\epsilon_{t-1}$ represents the forecasting error from the last prediction, whereas $\epsilon_t$ represents the forecasting we are making when predicting $y_t$. Of course, we can use more than one past term in a MA process. The amount of lags within a MA process is normally denoted as $Q$ (this will change to lower case q, as soon as we introduce the MA seasonal terms later in this post). Therefore, written in general form, a MA process has the following appearance:
$$y_t = \phi_0 + \sum^Q_{q=1} \theta_q \epsilon_{t-q} + \epsilon_t$$
Through the weighting parameters $\theta$, we are then able to weight forecasting errors in the past differently strong. Sensibly that would be that the most recent forecasting error gets a higher weight allocated, compared to a forecasting error in the past.

## Autoregressive models
The main difference between the moving average model and an autoregressive model is that whereas the MA model is using past forecasting errors to come up with an estimate for $y_t$, an AR process is using lagged version of $y$ itself. A 1st order model, for example, would take the following form:
$$y_t = \phi_0 + \phi_1 y_{y-1} + \epsilon_t$$
The value of $\phi$ plays a crucial role here. For example, assuming that the $\phi_1$ would be equal to 0, then the prediction for $y_t$ would be nothing other than a white noise process, which is not predictable. If $\phi_1$ would be equal to 1, then we would face a random walk, again something we cannot predict. For the case $\phi_1$ being greater than 1, we would have an exploding process with infinite variance. Needless to say, an exploding process cannot be predicted. Given these reasons we normally restrict the parameter $\phi$ to be between $-1 < \phi < 1$ in order for the time series to be stationary.

Lastly, for the generalized case we the P to denote the number of AR terms used in our model (a notation that will also change when introducing seasonal AR terms).
$$y_t = \phi_0 + \sum^P_{p=1} \phi_p y_{t-p} + \epsilon_t$$

## ARMA
When putting the two classes from above together, we end up with a ARMA model, which in its general form looks like the following:
$$y_t = \phi_0 + \sum^P_{p=1} y_{t-p} + \sum^Q_{q=1} \theta_q \epsilon_{t-q} + \epsilon_t$$

## Stationarity and ARIMA
When it comes to time series modelling, we care a lot about whether our time is actually predictable. If we for exampel face a model that continuously increasing over time (i.e. has an upward trend), we would find a positive and significant correlation between that time series and pretty much every other time series which is also facing upward. The price of cheese over the years would then be perfectly explained by the average climate in the arctics. That is, of course, non-sense and only the result of spurious correlation.

In order to cope with this data we therefore need to adjust the target variable to shrink the effect of these outliers. For that, we have multiple options at hand.

1. Dropping
2. Winsorizing
3. Transforming

Firstly, we could simply drop observations which exceed a pre-defined threshold. This, of course has several drawbacks. First, who says what the threshold is. Second, if we remove those datapoints but they appear in the test data, then we have the problem that our model never saw those spikes.

Secondly, we could [winsorize](https://en.wikipedia.org/wiki/Winsorizing) the data. Winsorizing describes the process of limiting the data to certain self-set threshold. One could for example assign all observations in the top 1 percent the value of the 99th percentile. That method would dampen the effect of the outliers. This method has the problem as the dropping method, one would have to somewhat subjectively decide what a suitable cutoff value would be.

Thirdly, and also the method we chose is to transform the data altogether. A common method when dealing with heaviliy skewed data would be to apply the natural logarithm to the data. That would dampen the effect of very high values and squeeze the entire dataset altogether. One could then use that transformed data for the predictions, before transforming the values back into their original format. Alternative to taking the natural logarithm would be the apply a so-called boxcox transformation.

### Boxcox Transformation
As the natural logarithmic function, the boxcox transformation represents a monotonic transformation of the data. That means that the order of the data is preseverd, even though the difference between datapoints is changed. That means that if a datapoint is higher than another one, then this will stay also after the transformation, only the distance between them changes.

## Box-Jenkins Approach 

When speciyfing the the SARIMAX, there is an overwhelming amount of hyper-parameters to choose. Next to specifying the number of Autogregressive (AR) and Moving Averages (MA), we also have to state the parameters for the seasonality. One approach would be to gridsearch over these hyper-parameters and then to go with the ones which resulted in the best mean absolute error. The most common and more academic approach for time series model selection is the so-called Box-Jenkins approach.

The Box-Jenkins method consists out of two steps:
- Visual inspection the Autocorrelation (ACF) function and Partial Autocorrelation function (PACF)
- Selection of potential models, fitting those models and assessing the serial autocorrelation of the residuals through the ljung-box test. If we find significant autocorrelation within the residuals, the procedure is repeated

The visual inspection of the ACF and PACF plots follow a certain rule, which is specified in the image below:
![title](img/boxjenkins.png)


When it comes to our time-series, we have to be aware that we are dealing with highly seasonal data. Therefore we do not only have to assess the correct number of lags for the nonseasonal dynamics (normally denoted as *p* and *q*), but also for the seasonal effects.

### ACF and PACF plots

The two plots below show us the autocorrelation and partial autocorrelation function of the de-trended target variable. From the ACF the seasonality is nicely visible. Given that we are dealing with weekly data, we can see spikes at 52 and 104 lags. When looking at the partial autocorrelation function we see that the first three lags contain significant individual information. Applying the Box-Jenkins rule to these two plots, it would make sense to include around up to three AR terms, as well as the same amount for the MA terms. When it comes to the seasonal terms, we find significant that the 52 lag is according to the ACF not significant, wheras the 104 lag, hence two years, is. Therefore, we are testing up to two lags for the seasonal ACF and PACF terms.

### Ljung Box Test and the importance of serial correlation
Now it is time to test all candidate models and assess the residual autocorrelation of each model afterwards. Luckily for us, `statsmodels` provides us this value. Like all outputs from statsmodels, it is not easily extractable, given that we find it deep inside a html table of the output, but the following code takes care of that extraction. 

```python
def sarimax_ljbox_parameters(target, exog, start_P, max_P, start_Q, max_Q, start_p,
                             max_p, start_q, max_q, seasonality, max_iter, trend):
    dict_lb_test = {}
    for P in range(0, max_P + 1):
        for Q in range(0, max_Q + 1):
            for p in range(0, max_p + 1):
                for q in range(0, max_q + 1):
                    mod = sm.tsa.statespace.SARIMAX(target, trend=trend, exog=feature_data,
                                                    order=(p, 0, q), seasonal_order=(P, 0, Q, seasonality),
                                                    enforce_stationarity=False, enforce_invertibility=False)
                    res = mod.fit(disp=False, max_iter=max_iter)
                    results_as_html = res.summary().tables[2].as_html()
                    results_df = pd.read_html(results_as_html, index_col=0)[0]

                    dict_lb_test[f"'P':{P}, 'Q':{Q}, 'p':{p}, 'q':{q}"] = {}
                    dict_lb_test[f"'P':{P}, 'Q':{Q}, 'p':{p}, 'q':{q}"]["#params"] = P + Q + p + q
                    dict_lb_test[f"'P':{P}, 'Q':{Q}, 'p':{p}, 'q':{q}"]["AIC"] = res.aic
                    dict_lb_test[f"'P':{P}, 'Q':{Q}, 'p':{p}, 'q':{q}"]["Q"] = results_df.iloc[1, 0]                    
                    print(f"P:{P}, Q:{Q}, p:{p}, q:{q} - Done")
    return dict_lb_test
```

```python
seasonality = 52
max_iter = 1_000
trend = "ct"
start_P = start_Q = 0
max_P = max_Q = 1
start_p = start_q = 0
max_p = 3
max_q = 1

file_name = f"{INTERIM_PATH}/dict_lb_sj.pickle"
if not os.path.isfile(file_name):
    dict_lb_sj = sarimax_ljbox_parameters(series_transformed_target, feature_data, start_P, max_P, start_Q,
                                          max_Q, start_p, max_p, start_q, max_q, seasonality, max_iter, trend)
    with open(file_name, 'wb') as handle:
        pickle.dump(dict_lb_sj, handle, protocol=pickle.HIGHEST_PROTOCOL)
else:
     dict_lb_sj = pickle.load(open(file_name, "rb"))
```

### Result assessment
After hyper-parameterizing the model, it is now time to assess how well the model performs. This is normally done through a cross-validation which we import from scikit-learn. Exactly that is the problem for our SARIMAX model, since it comes from statsmodels and not from scikit. For that reason we cannot simply insert our model into the `cross_validation` function. There is 



```python
class SARIMAXWrapper(BaseEstimator, RegressorMixin):
    """This class is a wrapper for a statsmodel model, in order to use the model within the sklearn
    cross validation."""

    def __init__(self, P, Q, p, q, trend, max_iter, seasonality, lambda_value):
        self.P = P
        self.Q = Q
        self.p = p
        self.q = q
        self.max_iter = max_iter
        self.trend = trend
        self.seasonality = seasonality
        self.lambda_value = lambda_value
        
    def inverse_boxcox(self, series, index):
        inverse_box_array = inv_boxcox(series.values, self.lambda_value)
        prediction_series = round(pd.Series(inverse_box_array)).astype(int) - 1 # Minus one because it added before 
        prediction_series.index = index
        return prediction_series

    def fit(self, X, y, verbose=False):
        model_ = sm.tsa.statespace.SARIMAX(y, trend=self.trend, exog=X, order=(self.p, 0, self.q),
                                           seasonal_order=(self.P, 0, self.Q, self.seasonality),
                                           enforce_stationarity=False, enforce_invertibility=False)
        self.results_ = model_.fit(disp=False, max_iter=self.max_iter)

        if verbose:
            self.results_.plot_diagnostics(figsize=(15, 15))
    
        return self

    def predict(self, X, in_sample=False, inverse_transform=False):
        if in_sample:
            predict_dy = self.results_.predict()
        else:
            predict_dy = self.results_.get_forecast(steps=len(X), exog=X).predicted_mean

        if inverse_transform:
            predict_dy = self.inverse_boxcox(predict_dy, X.index)
        return predict_dy
```

This class is now ready to be put into the `cross_val_score` from sklearn.

```python
N = 3
N_SPLITS = 5
time_split = TimeSeriesSplit(n_splits=N_SPLITS)
df_cross_val_scores = pd.DataFrame(index=list(range(N * N_SPLITS)), columns=["Model", "NegMeanAbsError"])

for i in range(N):
    key_values = list(sorted_dict_lb_sj.items())[i][0]
    dict_lags = ast.literal_eval(f"{{{key_values}}}")
    combined_dict = {**dict_lags, **dict_standard_params}
    sx = SARIMAXWrapper(**combined_dict)
    cv_scores = cross_val_score(sx, feature_data, series_transformed_target, cv=time_split)

    range_begin, range_end = (i * N_SPLITS), (((i+1) * N_SPLITS) - 1)
    df_cross_val_scores.loc[range_begin:range_end, "Model"] = f"{{{key_values}}}"
    df_cross_val_scores.loc[range_begin:range_end, "NegMeanAbsError"] = cv_scores
```
